```python
import pandas as pd
import datetime
import numpy as np
```

```python
# xls to csv
train = pd.read_excel(u'107年 竹苗空品區/107年新竹站_20190315.xls') #  index_col=0
# xls.to_csv('107年 竹苗空品區/107年新竹站_20190315.csv', encoding='big5')
# train = pd.read_csv('107年 竹苗空品區/107年新竹站_20190315.csv', encoding='big5', index_col = False)
```

```python
month_slice = train['日期'].apply(lambda x: int(x[5:7]))
# Truth value of a Series is ambiguous. Use a.empty, a.bool(), a.item(), a.any() or a.all()
df_train = train[(month_slice < 12) & (month_slice > 9)].reset_index(drop=True)
df_test = train[ month_slice == 12].reset_index(drop=True)
```

### 這邊必須把nr先轉成０跟填補空值

```python
df_train = df_train.drop(columns=['日期', '測站', '測項'], axis=1)
df_test = df_test.drop(columns=['日期', '測站', '測項'], axis=1)
```

```python
df_train = df_train.replace({'NR':0})
df_test = df_test.replace({'NR':0})
```

```python
list(range(6))
```

```python
for col in range(df_train.shape[0]):
    for i in range(df_train.shape[1]):
        if isinstance(df_train.iloc[col, i], str):
            df_train.iloc[col, i] = np.nan
```

```python
for col in range(df_test.shape[0]):
    for i in range(df_test.shape[1]):
        if isinstance(df_test.iloc[col, i], str):
            df_test.iloc[col, i] = np.nan
```

```python
a = df_train[0:18]
for i in range(18, len(df_train), 18):
    a = np.concatenate((a, df_train[i:i+18]), axis=1)
```

```python
b = df_test[0:18]
for i in range(18, len(df_test), 18):
    b = np.concatenate((b, df_test[i:i+18]), axis=1)
```

```python
test_item = list(train['測項'][0:18])
time_list = list(np.arange('2018-10-01', '2018-12-01', dtype='datetime64[h]'))
df_train =  pd.DataFrame(data=a, index=test_item, columns=time_list)
test_item = list(train['測項'][0:18])
time_list = list(np.arange('2018-12-01', '2019-01-01', dtype='datetime64[h]'))
df_test =  pd.DataFrame(data=b, index=test_item, columns=time_list)
# df_test =  pd.DataFrame(data=b, columns=time_list)
```

```python
df_train = (df_train.ffill(axis=1) + df_train.bfill(axis=1)) /2
df_test = (df_test.ffill(axis=1) + df_test.bfill(axis=1)) /2
```

```python
df_test.isnull().sum(axis=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y_train = []
for i in range(6,len(df_train.iloc[0])):
    y_train.append(df_train.loc['PM2.5'][i])
```

```python
y_test = []
for i in range(6,len(df_test.iloc[0])):
    y_test.append(df_test.loc['PM2.5'][i])
```

```python
x_train=[]
for i in range(0,len(df_train.iloc[0])-6):
    for j in range(18):
        x_train.append(df_train.iloc[:, i:i+6].iloc[j])
x_test=[]
for i in range(0,len(df_test.iloc[0])-6):
    for j in range(18):
        x_test.append(df_test.iloc[:, i:i+6].iloc[j])
```

```python
x_train = np.array(x_train).reshape(-1, 108)
x_test = np.array(x_test).reshape(-1, 108)
```

```python
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
lr = LinearRegression(normalize=True)
lr.fit(x_train, y_train)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print("MSE: %.2f"% mean_squared_error(pred, y_test))
print("MAE:  %.2f" %mean_absolute_error(pred, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100)
rf.fit(x_train, y_train)
pred = rf.predict(x_test)
```

```python
print("MSE: %.2f"% mean_squared_error(pred, y_test))
print("MAE:  %.2f" %mean_absolute_error(pred, y_test))
```

# 加深度

```python
rf = RandomForestRegressor(n_estimators=250, max_depth=5)
rf.fit(x_train, y_train)
pred = rf.predict(x_test)
```

```python
print("MSE: %.2f"% mean_squared_error(pred, y_test))
print("MAE:  %.2f" %mean_absolute_error(pred, y_test))
```

# min-max 標準化

```python
def max_min_normalization(x, max, min):
    return (x - min) / (max - min)
```

```python
x_train_mm = max_min_normalization(x_train, np.max(x_train), np.min(x_train))
x_test_mm = max_min_normalization(x_test, np.max(x_test), np.min(x_test))
y_train_mm = max_min_normalization(y_train, np.max(y_train), np.min(y_train))
y_test_mm = max_min_normalization(y_test, np.max(y_test), np.min(y_test))
```

```python
rf = RandomForestRegressor(n_estimators=250, max_depth=5)
rf.fit(x_train_mm, y_train_mm)
pred = rf.predict(x_test_mm)
```

```python
print("MSE: %.4f"% mean_squared_error(pred, y_test_mm))
print("MAE:  %.4f" %mean_absolute_error(pred, y_test_mm))
```
